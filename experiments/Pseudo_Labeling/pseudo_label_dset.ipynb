````python
import pandas as pd
import json
from pycocotools.coco import COCO
````

````python
df_submissioned = pd.read_csv('output.csv')
df_submissioned[["image_id","PredictionString"]]
````

````python
img_ids = []
file_names = [] # 순서대로 중복이 계속해서 존재하는 이미지 id들
obj_labels = []
obj_scores = []
obj_xmins = []
obj_ymins = []
obj_ws = []
obj_hs = []

for id, string in zip(df_submissioned["image_id"],df_submissioned["PredictionString"]): # 이미지 개수만큼의 loop
    splited_str = string.split(' ')[:-1] # 마지막에 스페이스가 존재하므로 삭제해서 -1
    # len(splited_str[0::6]) # obj의 개수
    for i in range(len(splited_str[0::6])):
        img_ids.append(id.split("/")[1].split(".")[0])
        file_names.append(id)
    obj_labels += splited_str[0::6]
    obj_scores += splited_str[1::6]
    obj_xmins += splited_str[2::6]
    obj_ymins += splited_str[3::6]
    obj_ws += [ float(row[0])-float(row[1]) for row  in zip(splited_str[4::6], splited_str[2::6]) ]
    obj_hs += [ float(row[0])-float(row[1]) for row  in zip(splited_str[5::6], splited_str[3::6]) ]
````

````python
# json["images"] 만들기
widths = [1024] * len(img_ids)
heights = [1024] * len(img_ids)
file_names
licenses = [0] * len(img_ids)
filcker_urls = [None] * len(img_ids)
coco_urls = [None] * len(img_ids)
date_captured = ["2020-12-23 16:20:30"] * len(img_ids)
file_names

df_img = pd.DataFrame()
df_img["width"] = widths
df_img["height"] = heights
df_img["file_name"] = file_names
df_img["license"] = licenses
df_img["flickr_url"] = filcker_urls
df_img["coco_url"] = coco_urls
df_img["date_captured"] = date_captured
df_img["id"] = img_ids
````

````python
print(len(df_img))
df_images = df_img.drop_duplicates()
print(len(df_images)) # 이미지의 총 개수 4871개 정상 확인
````

````python
df_images = df_images.astype({'id':'int'}) # id int type 변환
````

````python
df_images['id'] = df_images['id'] + 4883
````

````python
pred_images = df_images.to_json(orient='records', indent=4) # 생성완료
````

````python
# 양식에 맞도록 자리수 변환
img_ids_for_obj = [int(i) + 4883 for i in img_ids]
obj_scores = [i for i in obj_scores]
obj_xmins = [round(float(i),1) for i in obj_xmins] # str to float
obj_ymins = [round(float(i),1) for i in obj_ymins]
obj_ws = [round(float(i),1) for i in obj_ws]
obj_hs = [round(float(i),1) for i in obj_hs]
obj_area = [round(i[0]*i[1],2) for i in zip(obj_ws, obj_hs)] # 
obj_ids = [ i+23144 for i in range(len(img_ids_for_obj)) ]
obj_labels = [int(i) for i in obj_labels]
obj_bbox = []

for idx in range(len(obj_ids)):
    obj_bbox.append([obj_xmins[idx], obj_ymins[idx], obj_ws[idx], obj_hs[idx]])
````

````python
len(obj_bbox), len(img_ids_for_obj), len(obj_labels)
````

### annotation df 생성

````python
df_annotations = pd.DataFrame()
df_annotations["image_id"] = img_ids_for_obj
df_annotations["category_id"] = obj_labels
df_annotations["area"] = obj_area
df_annotations["bbox"] = obj_bbox
df_annotations["iscrowd"] =  [0 for i in range(len(obj_ids))]
df_annotations["id"] = obj_ids
df_annotations["confidence"] = [float(i) for i in obj_scores]
````

````python
df_annotations # train은 23,000개 preds는 137,121개
````

````python
confi_threshold = 0.5 # label로 판단할 기준
df_annotations = df_annotations[df_annotations["confidence"]>confi_threshold]
df_annotations = df_annotations.drop("confidence", axis=1) # confidence 삭제
````
