```python
import json
import numpy as np
from sklearn.model_selection import StratifiedGroupKFold

# load json
annotation = "/opt/level2_objectdetection_cv-level2-cv-04/folded_anns/train.json" # 개인 train.json 파일 위치 입력하세요

with open(annotation) as f:
    data = json.load(f)
    
var = [(ann['image_id'], ann['category_id']) for ann in data['annotations']] ## 이미지 id와 카테고리 id만 가져옴

X = np.ones((len(data['annotations']),1)) # annotations의 길이만큼 1을 생성함
y = np.array([v[1] for v in var]) # 카테고리의 id가 담긴 np를 생성함.
groups = np.array([v[0] for v in var]) # img_id가 담긴 np를 생성함.

cv = StratifiedGroupKFold(n_splits=5, shuffle=True, random_state=411)

train_indices = list()
val_indices = list()
for train_idx, val_idx in cv.split(X, y, groups):
    train_indices.append(train_idx)
    val_indices.append(val_idx)

print(len(groups[train_idx]))
print(len(y[train_idx]))
```

```python
num_of_img = set([i[0] for i in var])
print('num of obj:',len(data['annotations']))
print('num of img:', len(num_of_img))
```

```python
# 뽑힌 annots를 다 가지고 옴
train_annots = []
val_annots = []
for train_idx, valid_idx in zip(train_indices, val_indices):
    train_annot = np.array(data['annotations'])[train_idx].tolist()
    val_annot = np.array(data['annotations'])[valid_idx].tolist()
    train_annots.append(train_annot)
    val_annots.append(val_annot)
```

```python
train_img_indices = []
for train_annot in train_annots:
    ids = list()
    for d in train_annot:
        ids.append(d['image_id'])
    train_img_indices.append(list(set(ids)))

val_img_indices = []
for val_annot in val_annots:
    ids = list()
    for d in val_annot:
        ids.append(d['image_id'])
    val_img_indices.append(list(set(ids)))
```

```python
train_imgs = []
for img_indice in train_img_indices:
    train_img = np.array(data['images'])[img_indice].tolist()
    train_imgs.append(train_img)

val_imgs = []
for img_indice in val_img_indices:
    val_img = np.array(data['images'])[img_indice].tolist()
    val_imgs.append(val_img)
```
